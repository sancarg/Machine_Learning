## Logistic Model

### Vanilla Logistic Model

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
operations = [("transformer", column_trans), ("log", LogisticRegression(class_weight='balanced',max_iter=10000,
                                                                          random_state=101))]

pipe_log_model = Pipeline(steps=operations)
```

```python
pipe_log_model.fit(X_train,y_train)
eval_metric(pipe_log_model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.metrics import make_scorer, f1_score, precision_score, recall_score

classes = ["Black", "Hispanic", "White"]

# create a dictionary to hold all scorers
scoring = {}

for cls in classes:
    scoring[f"f1_{cls}"] = make_scorer(f1_score, average=None, labels=[cls])
    scoring[f"precision_{cls}"] = make_scorer(precision_score, average=None, labels=[cls])
    scoring[f"recall_{cls}"] = make_scorer(recall_score, average=None, labels=[cls])
#make_scorer expects a single scalar, not an array.
# add global metrics
scoring["f1_macro"] = make_scorer(f1_score, average="macro")
scoring["f1_weighted"] = make_scorer(f1_score, average="weighted")
```

```python
operations = [("transformer", column_trans), ("log", LogisticRegression(class_weight='balanced',max_iter=10000,
                                                                          random_state=101))]

model = Pipeline(steps=operations)

scores = cross_validate(model, X_train, y_train, scoring = scoring, cv = 10)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]
```

```python
param_grid = {
    "log__C": [0.001, 0.01, 0.1, 1, 10],
    "log__penalty": ["l1", "l2"],
    "log__solver": ["liblinear", "lbfgs"],
    "log__multi_class": ["ovr", "multinomial"],
}
```

```python
operations = [("OneHotEncoder", column_trans), ("log", LogisticRegression(class_weight='balanced',max_iter=10000,
                                                                          random_state=101))]

model = Pipeline(steps=operations)

log_model_grid = GridSearchCV(model,
                              param_grid,
                              verbose=3,
                              scoring=scoring,
                              refit="recall_Hispanic",   
                              n_jobs=-1, cv=5)
```

```python
log_model_grid.fit(X_train,y_train)
```

```python
log_model_grid.best_params_
```

```python
log_model_grid.best_score_
```

```python
# Get predictions
y_pred = log_model_grid.predict(X_test)

eval_metric(log_model_grid, X_train, y_train, X_test, y_test)

# --- Confusion Matrix (multiclass) ---
cm = confusion_matrix(y_test, y_pred, labels=["Black", "Hispanic", "White"])
print("\nConfusion Matrix (rows=true, cols=pred):\n")
print(cm)

# --- Heatmap ---
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Black", "Hispanic", "White"],
            yticklabels=["Black", "Hispanic", "White"])
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.title("Confusion Matrix (3 Classes)")
plt.show()
```

- Calculate the mean absolute SHAP value for each feature
- For multi-class models, shap_values is a list of arrays. We need to average them.
- The `np.abs(shap_values[0])` part is for the first class, `[0]` refers to the first class.
- It's better to average across all classes if you're doing multi-class classification.
- For simplicity and to match the single-target focus of your refit metric, we'll calculate the mean of the absolute values for the 'Hispanic' class.
- You will need to know which index corresponds to 'Hispanic'. Let's assume it's 0 for this example.

```python
shap_importance = np.mean(np.abs(shap_values[0]), axis=0)
```

```python
# 2. Get the feature names and their importance scores
feature_importance_df = pd.DataFrame({
    'feature': features,
    'importance': shap_importance
})
```

```python
# 3. Sort the features by importance in descending order
feature_importance_df = feature_importance_df.sort_values(
    by='importance', ascending=False
)
```

```python
# 4. Get the names of the top N features
# Let's say you chose max_display=20 for your plot.
num_top_features = 40
top_features_df = feature_importance_df.head(num_top_features)
top_feature_names = top_features_df['feature'].tolist()

print("Top", num_top_features, "most important features:")
top_features_df
```
