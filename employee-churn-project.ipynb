```python
# Checking for for duplicate rows
def duplicate_values(df):
    print("Duplicate check")
    print("There are", df.duplicated(subset=None, keep='first').sum(), "duplicated observations in the dataset.")
    
    duplicate_values = df.duplicated(subset=None, keep='first').sum()
    
    if duplicate_values > 0:
        df.drop_duplicates(keep='first', inplace=True)
        print(duplicate_values, " Duplicates were dropped!\n")
```

```python
# To view summary information about the columns
def first_look(col):
    print("column name    : ", col)
    print("--------------------------------")
    print("Per_of_Nulls   : ", "%", round(df[col].isnull().sum() / df.shape[0]*100, 2))
    print("Num_of_Nulls   : ", df[col].isnull().sum())
    print("Num_of_Uniques : ", df[col].nunique())
    print("Duplicates     : ", df.duplicated(subset = None, keep = 'first').sum())
    print(df[col].value_counts(dropna = False))
    
```

```python
# Evaluating the model's performance by classification report and confusion matrix  
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)    
    print("TEST REPORT")
    print(classification_report(y_test, y_pred_test))
    confusion_matrix_test = confusion_matrix(y_test, y_pred_test)
    ConfusionMatrixDisplay(confusion_matrix_test).plot()
    print("\n"*3, "-*"*30)
    print("TRAIN REPORT")
    print(classification_report(y_train, y_pred_train))
    confusion_matrix_train = confusion_matrix(y_train, y_pred_train)
    ConfusionMatrixDisplay(confusion_matrix_train).plot()
    plt.show();
```

<h2 style="color:white;text-align: left;font-size: 24px;background-color: blueviolet;padding: 10.5px">Scaling</h2>

Scaling
Some machine learning algorithms are sensitive to feature scaling while others are virtually invariant to it. Machine learning algorithms like linear regression, logistic regression, neural network, etc. that use gradient descent as an optimization technique require data to be scaled. Also distance algorithms like KNN, K-means, and SVM are most affected by the range of features. This is because behind the scenes they are using distances between data points to determine their similarity.

Scaling Types:

Normalization: Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max scaling.

Standardization: Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero and the resultant distribution has a unit standard deviation.

Hopkins Test
- The Hopkins statistic (Lawson and Jurs 1990) is used to assess the clustering tendency of a data set by measuring the probability that a given data set is generated by uniform data distribution. In other words, it tests the spatial randomness of the data. The Hopkins Test explains that you may expect 3 different results: 
  1. H = 0.5 (the dataset reveals no clustering structure)  
  2. H close to 1.0, a significant evidence that the data might be cluster-able. 
  3. H is close to 0, in this case the test is indecisive (data is randomly distributed, so it is neither clustered nor random)

```python
from sklearn.neighbors import BallTree
def hopkins(data_frame, sampling_size):
    

    if type(data_frame) == np.ndarray:
        data_frame = pd.DataFrame(data_frame)

    # Sample n observations from D : P

    if sampling_size > data_frame.shape[0]:
        raise Exception(
            'The number of sample of sample is bigger than the shape of D')

    data_frame_sample = data_frame.sample(n=sampling_size)

    # Get the distance to their neirest neighbors in D : X

    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(data_frame_sample, k=2)
    data_frame_sample_distances_to_nearest_neighbours = dist[:, 1]

    # Randomly simulate n points with the same variation as in D : Q.

    max_data_frame = data_frame.max()
    min_data_frame = data_frame.min()

    uniformly_selected_values_0 = np.random.uniform(min_data_frame[0], max_data_frame[0], sampling_size)
    uniformly_selected_values_1 = np.random.uniform(min_data_frame[1], max_data_frame[1], sampling_size)

    uniformly_selected_observations = np.column_stack((uniformly_selected_values_0, uniformly_selected_values_1))
    if len(max_data_frame) >= 2:
        for i in range(2, len(max_data_frame)):
            uniformly_selected_values_i = np.random.uniform(min_data_frame[i], max_data_frame[i], sampling_size)
            to_stack = (uniformly_selected_observations, uniformly_selected_values_i)
            uniformly_selected_observations = np.column_stack(to_stack)

    uniformly_selected_observations_df = pd.DataFrame(uniformly_selected_observations)

    # Get the distance to their neirest neighbors in D : Y

    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(uniformly_selected_observations_df, k=1)
    uniformly_df_distances_to_nearest_neighbours = dist

    # return the hopkins score

    x = sum(data_frame_sample_distances_to_nearest_neighbours)
    y = sum(uniformly_df_distances_to_nearest_neighbours)

    if x + y == 0:
        raise Exception('The denominator of the hopkins statistics is null')

    return x / (x + y)[0]
scalers = [StandardScaler(), MinMaxScaler()]
# Loop through scalers
for scaler in scalers:
    # Scale the data using the current scaler
    scaled_data = scaler.fit_transform(df_enc)
    
    # Compute the Hopkins score
    score = hopkins(scaled_data, scaled_data.shape[0])
    
    print(f'{scaler}:', score)
```

<span style="font-size:20px;color:purple;font-family:cursive"> Both Hopkins statistics are relatively low, indicating a low clustering tendency in our dataset. However, the difference between the two values is quite small. The choice between StandardScaler and MinMaxScaler may not be crucial for the clustering task. We have prefered MinMaxScaler because of beter results. </span>

<h2 style="color:white;text-align: left;font-size: 24px;background-color: blueviolet;padding: 10.5px;">The Elbow Method</h2>

- "Elbow Method" can be used to find the optimum number of clusters in cluster analysis. The elbow method is used to determine the optimal number of clusters in k-means clustering. The elbow method plots the value of the cost function produced by different values of k. If k increases, average distortion will decrease, each cluster will have fewer constituent instances, and the instances will be closer to their respective centroids. However, the improvements in average distortion will decline as k increases. The value of k at which improvement in distortion declines the most is called the elbow, at which we should stop dividing the data into further clusters.

    [The Elbow Method](https://en.wikipedia.org/wiki/Elbow_method_(clustering)

    [The Elbow Method2](https://medium.com/@mudgalvivek2911/machine-learning-clustering-elbow-method-4e8c2b404a5d)

    [KMeans](https://towardsdatascience.com/understanding-k-means-clustering-in-machine-learning-6a6e67336aa1)

<h3 style="color:white;text-align: left;font-size: 24px;background-color: blueviolet;padding: 8.5px;">Feature Importance With SHAP</h3>

- SHAP (SHapley Additive exPlanations) is a game theoretic approach to explain the output of any machine learning model.
- When assessing feature importance, it's essential to consider both "error contribution" and "prediction contribution." These two aspects provide a more comprehensive understanding of how features impact your machine learning model's predictions.
- When it comes to accounting for error contribution in your machine learning model, SHAP (SHapley Additive exPlanations) values are generally a more comprehensive and informative tool compared to feature importance scores.

[Explanatory article-1 for "SHAP"](https://towardsdatascience.com/using-shap-values-to-explain-how-your-machine-learning-model-works-732b3f40e137)

[Explanatory article-2 for "SHAP"](https://towardsdatascience.com/your-features-are-important-it-doesnt-mean-they-are-good-ff468ae2e3d4)

[Explanatory article-3 for "SHAP"](https://medium.com/dataman-in-ai/explain-your-model-with-the-shap-values-bc36aac4de3d)

```python
explainer = shap.TreeExplainer(rfc_S_model.named_steps['rfc_S'])
shap_values = explainer.shap_values(X_test_transformed)
shap_obj = explainer(X_test_transformed)
shap.plots.beeswarm(shap_obj[:,:,1], max_display=18) # only for one dimension
                                                     # max_display default is 10
#RandomForest is a bit special, when something fails with the plots, try to add Explanation Object.
```

- SHAP provides numeric importance values for every feature for every individual prediction.  The beeswarm plots every one of these values as a dot on a chart.  The horizontal axis indicates how predictive that feature is for that individual in a positive or negative direction.  In addition, the beeswarm chart adds color coding of each dot based on how the feature value for that individual compares to the average for the entire population. Color indicates whether that variable is high (red) or low (blue) for that observation.
- We can also see correlation between each feature and the target variable. A high level of the time_spend_company has a high and positive impact on target. Note the positive impact is shown on the X-axis. Similarly, we can say that the high value of satisfaction_level is negatively correlated with the target variable.

<h4 style="color:white;text-align: left;font-size: 24px;background-color: blueviolet;padding: 8.5px;">Force Plot</h4>

- The SHAP force plot and waterfall plot are similar in that they both show how the features of a data point contribute to the models prediction, as they both show the magnitude and direction of the contribution as arrows or bars. 
- The main difference between the two plots is the orientation. SHAP force plots show the features from left to right, with the positive contributions on the left and the negative contributions on the right. Waterfall plots show the features from top to bottom, with the positive contributions at the top and the negative contributions at the bottom.

```python
compare = pd.DataFrame({"Model": ["KNN",
                                  "Random Forest",
                                  "Random_Forest_Feat_Imp",
                                  "XGBoost",
                                  "XGBoost_Feat_Imp"
                                 ],
                       "Precision": [KNN_grid_precision, rf_grid_model_precision, rfc_FI_model2_precision, xgb_grid_precision, xgb_FI_model2_precision],
                       "Recall": [KNN_grid_recall, rf_grid_model_recall, rfc_FI_model2_recall, xgb_grid_recall, xgb_FI_model2_recall],
                       "F1": [KNN_grid_f1, rf_grid_model_f1, rfc_FI_model2_f1, xgb_grid_f1, xgb_FI_model2_f1],
                       "PR_AUC": [KNN_grid_precision_recall, rf_grid_model_precision_recall, rfc_FI_model2_precision_recall, xgb_grid_precision_recall, xgb_FI_model2_precision_recall]})

def labels(ax):
    ax.bar_label(ax.containers[0],fmt="%.3f")

plt.figure(figsize=(14,10))
plt.subplot(411)
compare = compare.sort_values(by="Precision", ascending=False)
ax=sns.barplot(x="Precision", y="Model", data=compare, palette="plasma")
labels(ax)

plt.figure(figsize=(14,10))
plt.subplot(412)
compare = compare.sort_values(by="Recall", ascending=False)
ax=sns.barplot(x="Recall", y="Model", data=compare, palette="gist_rainbow_r")
labels(ax)

plt.figure(figsize=(14,10))
plt.subplot(413)
compare = compare.sort_values(by="F1", ascending=False)
ax=sns.barplot(x="F1", y="Model", data=compare, palette="gnuplot")
labels(ax)

plt.figure(figsize=(14,10))
plt.subplot(414)
compare = compare.sort_values(by="PR_AUC", ascending=False)
ax=sns.barplot(x="PR_AUC", y="Model", data=compare, palette="tab20b_r")
labels(ax)
plt.show()
```

- Recall is particularly important in imbalanced data scenarios when the cost of false negatives (missed instances of the minority class) is high. If we lose the employee, will have to pay all the costs of acquiring a replacement employee as cost of recruiting, advertising and etc.
- So we will prefer XGBOOST Model (xgb_FI_model).
